```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("contohdataclustering.xlsx")
```

```python
df = df.drop(['No'], axis=1)
```

```python

datax = df['x']
datay = df['y']
df.head(40)
```

```python
k = 3;
x = [];
centerx = []
centery = []
for i in range(k):
    x.append(np.random.randint(30))
    centerx.append(datax[x[i]])
    centery.append(datay[x[i]])



```

```python

precenterx = []
precentery = []
precenterx.append(0)
precenterx.append(0)
precenterx.append(0)
precentery.append(0)
precentery.append(0)
precentery.append(0)

while True:
    titik1 = []
    titik2 = []
    titik3 = []
    precenterx[0] = centerx[0]
    precenterx[1] = centerx[1]
    precenterx[2] = centerx[2]
    precentery[0] = centery[0]
    precentery[1] = centery[1]
    precentery[2] = centery[2]
    jumx1 = 0
    jumx2 = 0
    jumx3 = 0
    jumy1 = 0
    jumy2 = 0
    jumy3 = 0
    for i in range(len(datax)):
        distance1 = np.sqrt(((centerx[0] - datax[i])**2) + ((centery[0] - datay[i])**2))
        distance2 = np.sqrt(((centerx[1] - datax[i])**2) + ((centery[1] - datay[i])**2))
        distance3 = np.sqrt(((centerx[2] - datax[i])**2) + ((centery[2] - datay[i])**2))
        if ((distance1 < distance2) and (distance1 < distance3)):
            titik1.append(i)
        elif ((distance2 < distance1) and (distance2 < distance3)):
            titik2.append(i)
        elif ((distance3 < distance1) and (distance3 < distance2)):
            titik3.append(i)

    for i in range(len(titik1)):
        jumx1 = jumx1 + datax[titik1[i]]
    averagex1 = jumx1/len(titik1);
    centerx[0] = averagex1;

    for i in range(len(titik1)):
        jumy1 = jumy1 + datay[titik1[i]]
    averagey1 = jumy1/len(titik1);
    centery[0] = averagey1;

    for i in range(len(titik2)):
        jumx2 = jumx2 + datax[titik2[i]]
    averagex2 = jumx2/len(titik2);
    centerx[1] = averagex2;

    for i in range(len(titik2)):
        jumy2 = jumy2 + datay[titik2[i]]
    averagey2 = jumy2/len(titik2);
    centery[1] = averagey2;

    for i in range(len(titik3)):
        jumx3 = jumx3 + datax[titik3[i]]
    averagex3 = jumx3/len(titik3);
    centerx[2] = averagex3;

    for i in range(len(titik3)):
        jumy3 = jumy3 + datay[titik3[i]]
    averagey3 = jumy3/len(titik3);
    centery[2] = averagey3;


    print(centerx)
    print(centery)
    if((precenterx[0] == centerx[0]) and (precenterx[1] == centerx[1]) and (precenterx[2] == centerx[2]) and (precentery[0] == centery[0]) and (precentery[1] == centery[1]) and (precentery[2] == centery[2])):
        break




        
```

```python
plt.plot(centerx,centery, "ko")
plt.plot(datax[titik1],datay[titik1], "ro")
plt.plot(datax[titik2],datay[titik2], "bo")
plt.plot(datax[titik3],datay[titik3], "go")
plt.show
```
